```python
def get_validation_details(row, mapping_dict):
    failed_validations = []
    for column in mapping_dict.keys():
        if row[column] == 0:
            failed_validations.append(mapping_dict[column])
    return ', '.join(failed_validations)

df['Julat_Details'] = df.apply(lambda row: get_validation_details(row, julat_mapping), axis=1)
df['Validation_Details'] = df.apply(lambda row: get_validation_details(row, konsistensi_mapping), axis=1)
```
